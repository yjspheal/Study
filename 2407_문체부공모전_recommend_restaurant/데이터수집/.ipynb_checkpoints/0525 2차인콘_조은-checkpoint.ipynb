## 크롤링

```python
import pandas as pd
import openpyxl
import os
import sys
import time
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from pyparsing import col
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from tqdm import tqdm 
import re
from selenium.webdriver.support.ui import WebDriverWait
warnings.filterwarnings('ignore')

#변수 지정
items = df_10['사업장명']
print(items)

#검색할 식당 데이터와 url을 담을 데이터 프레임 생성
df = pd.DataFrame(columns=['name', 'naverURL'])

#데이터 프레임이 잘 만들어졌는지 확인
df['name'] = items
df['naverURL']=''
df
```

```python
# 식당 url 얻기
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

res = driver.page_source  # 페이지 소스 가져오기
soup = BeautifulSoup(res, 'html.parser')  # html 파싱하여  가져온다

# 페이지 로딩 대기 시간
LOAD_WAIT_TIME = 3

# 각 식당의 URL 가져오기
for i, keyword in enumerate(df['name'].tolist()):
    # NaN 값인 경우 처리
    if pd.isna(keyword):
        continue
    
    # 네이버 지도 검색 URL 생성
    naver_map_search_url = f'https://map.naver.com/p/search/{keyword}'
    
    # 검색 URL로 이동하여 검색 실행
    driver.get(naver_map_search_url)
    time.sleep(LOAD_WAIT_TIME)
    
# 프레임 전환
    try:
        # 프레임으로 전환할 때까지 대기
        WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.TAG_NAME, 'iframe')))
        
        # 검색 결과가 있을 경우 첫 번째 식당 정보 클릭
        if len(driver.find_elements(By.CSS_SELECTOR, '#_pcmap_list_scroll_container > ul > li')) > 0:
            # 요소가 로드될 때까지 대기
            element = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, 'a.P7gyV'))
            )
            element.click()
            time.sleep(LOAD_WAIT_TIME)
            
            # 현재 페이지 URL에서 식당 코드 추출
            current_url = driver.current_url
            res_code = re.findall(r"place/(\d+)", current_url)
            if res_code:
                final_url = f'https://map.naver.com/p/search/{keyword}/place/{res_code[0]}'
                df.at[i, 'naverURL'] = final_url
                print(final_url)
        
    except Exception as e:
        print('Error:', e)

# WebDriver 종료
driver.quit()

# 식당명과 URL 출력
print(df)
```
