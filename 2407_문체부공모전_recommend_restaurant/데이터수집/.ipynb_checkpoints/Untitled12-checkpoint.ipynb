```python
import pandas as pd

# 기존 데이터 프레임
df = pd.DataFrame(columns=['지번주소', '도로명주소', '업체명', '업종', 'URL', 'ID'])

# DataFrame의 복사본 생성
df_copy = df.copy()

# 새로운 행 데이터
new_row_data = {'지번주소': '새로운 지번주소', '도로명주소': '새로운 도로명주소', '업체명': '새로운 업체명', '업종': '새로운 업종', 'URL': '새로운 URL', 'ID': '새로운 ID'}

# 새로운 행 추가
df_copy.loc[-1] = new_row_data
df_copy.index = df_copy.index + 1
df_copy = df_copy.sort_index()

# 결과 확인
print(df_copy)
```

```python
import pandas as pd

# 데이터프레임 불러오기
df = pd.read_csv('modified_대흥동.csv')

# 헤더 재설정
headers = ['Index', '지번주소', '도로명주소', '업체명', '업종', 'URL', 'ID']
df.columns = headers

# 인덱스 다시 설정
df.reset_index(drop=True, inplace=True)

# 수정된 데이터프레임 확인
print(df)


```

```python
# 수정된 데이터프레임을 새 파일로 저장
df.to_csv('modified_대흥동.csv', index=False)
```
