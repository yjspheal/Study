```python
import pandas as pd

df1 = pd.read_csv('../../2406_인사이콘2차_recommend_restaurant/final_df_0327_part1.csv')
df2 = pd.read_csv('../../2406_인사이콘2차_recommend_restaurant/final_df_0327_part2.csv')
df3 = pd.read_csv('../../2406_인사이콘2차_recommend_restaurant/final_df_0327_part3.csv')
df4 = pd.read_csv('./연남동_상세정보.csv')
```

```python
df1.drop(['Unnamed: 0.1','Unnamed: 0'], axis = 1, inplace = True)
df2.drop(['Unnamed: 0.1','Unnamed: 0'],  axis = 1, inplace = True)
df3.drop(['Unnamed: 0.1','Unnamed: 0'], axis = 1,  inplace = True)
df4.drop(['Unnamed: 0'], axis = 1,  inplace = True)
```

```python
result_df = pd.concat([df1, df2, df3, df4], axis=0)
```

```python
result_df = result_df.drop_duplicates().reset_index()
```

```python
result_df.to_csv('final_df_0713.csv', index = False, encoding= 'UTF-8-sig')
result_df.to_excel('final_df_0713.xlsx', index = False)
```

```python
mobum = pd.read_excel('서울특별시 마포구_모범음식점 목록.xlsx')
mobum
```

```python
# 포함 여부를 저장할 리스트를 초기화합니다.
contains_list = []

# result_df의 각 new_address 값을 반복합니다.
for new_address in result_df['new_address']:
    # new_address가 mobum['소재지(도로명)2']에 포함되는지 확인합니다.
    contains = mobum['소재지(도로명)2'].str.contains(new_address, regex=False).any()
    # 결과를 리스트에 추가합니다.
    contains_list.append(contains)

# 결과 리스트를 result_df에 새로운 열로 추가합니다.
result_df['contains'] = contains_list

# 포함된 행만 필터링하여 새로운 데이터프레임에 저장합니다.
result_df[result_df['contains']]
```
